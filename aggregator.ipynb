```python
df_numerical_qi = df[["Age", "ZIP"]]
print(df_numerical_qi.shape)
df_numerical_qi_dist = pdist(df_numerical_qi)
df_numerical_qi_dist_sq = squareform(df_numerical_qi_dist)
print(df_numerical_qi_dist_sq.shape)
```

```python
df_distances = pd.DataFrame(df_numerical_qi_dist_sq)
```

```python
df_distances
```

```python
df["Sum_of_Distances"] = df_distances.apply(lambda row: np.sum(row))
```

```python
df.head(4)
```

```python
average_idx = np.argmin(np.array(df["Sum_of_Distances"]))
```

```python
furthest_away_from_average_idx = np.argmax(np.array(df_distances[average_idx]))
```

```python
other_end = np.argmax(np.array(df_distances[furthest_away_from_average_idx]))
```

```python
other_end
```

```python
def make_k_group(seed_member):
    group = [seed_member]

    new_df = df_distances.drop(seed_member, axis = 0)
    for i in range(4):
        new_member = np.argmin(np.array(new_df[furthest_away_from_average_idx]))
        group.append(new_member)
        new_df = new_df.drop(new_member, axis = 0)
    print(group)
    return group
```

```python
make_k_group(35)
```
